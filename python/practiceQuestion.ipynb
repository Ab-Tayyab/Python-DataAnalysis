```python
# -21 function take a list of number and return largest and smalest one number 
def largest(a):
    n=a[0]
    m=a[0]
    for num in a:
        if(num>n):
            n=num
        if(m>num):
            m=num
    return n,m
print("Largest, smallest number is: ",largest([7,2,1,3,4,5,6]))
```

```python
# -22 grade system 
def grade(number):
    if(number>90):
        return "A+ grade"
    elif(number>80):
        return "B grade"
    elif(number>65):
        return "C grade"
    elif(number>50):
        return "D grade"
    else:
        return "Fail!"
print("You got ",grade(49))
```

```python
# -25 remove duplicate from list 
list = [1,3,2,2,4,5,7,6,7]
list2=[]
for i in range(len(list)):
    isFound = False
    for j in range(len(list2)):
        if(list[i]==list2[j]):
            isFound=True
            break
    if(isFound==False):
        list2.append(list[i])
print(list2)
```

```python
# -29 string reverse 
text = input("Enter string: ")
revStr=""
for i in text:
    revStr = i+revStr
print("Original string: ", text)
print("Reverse string: ", revStr)
# or 
revText = text[::-1]
print("Original string: ", text)
print("Reverse string: ", revText)
```
