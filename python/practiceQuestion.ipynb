```python
n1 = int(input("Enter first number: "))
n2 = int(input("Enter first number: "))
def sum(a,b):
    return a+b
def dif(a,b):
    return a-b
def pro(a,b):
    return a*b
def quo(a,b):
    a//b
print("Sum is: ",sum(n1,n2))
print("Difference is: ",dif(n1,n2))
print("Product is: ",pro(n1,n2))
print("Quotient is: ",quo(n1,n2))
```
