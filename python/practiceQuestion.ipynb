```python
# -21 function take a list of number and return largest and smalest one number 
def largest(a):
    n=a[0]
    m=a[0]
    for num in a:
        if(num>n):
            n=num
        if(m>num):
            m=num
    return n,m
print("Largest, smallest number is: ",largest([7,2,1,3,4,5,6]))
```

```python
# sort a list of tuples 2nd element using lambda 
elements = [(3,2),(1,4),(2,1),(4,3)]
n = len(elements)
for i in range(n):
    for j in range(n-1-i):
        if elements[j][1]>elements[j+1][1]:
            elements[j],elements[j+1]=elements[j+1],elements[j] #swap value
print("Using nested loops: ",elements)

# if you want sort using lambda 
sort_list = sorted(elements, key=lambda x:x[1])
print("Using lambda function: ",elements)
```

```python
import random

num = random.randint(1, 10)
tries = 3

while tries > 0:
    pick = int(input(f"Enter number between 1 to 10 (Attempts left: {tries}): "))

    if pick == num:
        print("🎉 You won!", pick)
        break
    elif pick < num:
        print("Your number is less than the target. Try a larger one!")
    else:
        print("Your number is greater than the target. Try a smaller one!")

    tries -= 1

else:
    print("❌ Game over! The correct number was:", num)
```

```python
# common elements in a two lists 
list1 = [1,2,3,4,5]
list2=[2,3,5,6,7]
common = []
for i in list1:
    if i in list2 and i not in common:
        common.append(i)
print(common)
```

```python
# unique elements in a two lists 
list1 = [1,2,3,4,5]
list2=[2,3,5,6,7]
common = []
for i in list1:
    if i not in list2 and i not in common:
        common.append(i)
for i in list2:
    if i not in list1 and i not in common:
        common.append(i)
print(common)
```
