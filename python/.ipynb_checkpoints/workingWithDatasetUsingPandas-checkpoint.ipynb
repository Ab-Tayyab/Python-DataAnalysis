```python
# handling data types and type converion 
df = pd.read_csv("./files/retail_sales.csv",
                parse_dates = ["Date"],
                dtype={
                    "Category": "category"
                })
df.info()
```

```python
# handling missing data
df = pd.read_csv(
    "./files/retail_sales.csv",
    parse_dates = ["Date"],
    dtype={
        "Category":"category",
        "Region":"category"
    }
)
df.head()
```

```python
df.info()
```

```python
print(df.isna().sum()) # detect missing values
```

```python
# fill missing age with median. we use statistic function mean, median, mode: for numerical column we use mean and median and for categorical value we use mode.
df['Sales']=df['Sales'].fillna(df['Sales'].median())
```

```python
# fill value with mode
df['Quantity'] = pd.to_numeric(df['Quantity'])
df['Quantity']=df['Quantity'].fillna(df['Quantity'].mode()[0])
```

```python
# fill with mode
df['Category']=df['Category'].fillna(df['Category'].mode()[0])
```

```python
df.info()
```

```python
print(df.isna().sum()) # detect missing values
```

```python
df.dropna(subset=['Region'],inplace=True)
```

```python
print(df.isna().sum()) # detect missing values
```

```python
# sotring and ranking data 
df = pd.read_csv(
    "./files/retail_sales.csv",
    parse_dates = ["Date"],
    dtype={
        "Category":"category",
        "Region":"category"
    }
)
df.head()
```

```python
# sort by age ascending, then salary descending
df_sorted = df.sort_values(by=['Quantity','Sales'],ascending=[False,True])
print("Sorted the Dataframe")
df_sorted.head()
```

```python
# rank salary within age groups
df['Sales_rank']=df.groupby('Quantity')['Sales'].rank(method='dense',ascending=False)
print("sales rank within quantity group \n",df)
```

```python
# Merging and joining Dataframes:  inner join, left join, right join, outer join.
#sample Dataframes
customers = pd.DataFrame({
    'CustomerID':[1,2,3],
    'Name':['Alice','Bob','Charlie']
})
orders = pd.DataFrame({
    'OrderID':[101,102,103,104],
    'CustomerID':[1,2,2,3],
    'Amount':[205,150,200,300]
})
# inner merge on customerID 
df_inner = pd.merge(customers,orders,on='CustomerID',how='inner')
print("Inner merge: \n",df_inner)
```

```python
# left merge
df_left = pd.merge(customers,orders,on='CustomerID',how='left')
print("Left merge: \n",df_left)
```

```python
# concatenate rows example
df_concat = pd.concat([customers,customers])
print("Left merge: \n",df_concat)
```

```python
# applying function-apply,map & pipe 
df = pd.read_csv("./files/ecommerce_sales_analytics_5000.csv",parse_dates=["order_date"])
df.info()
```

```python
# map(), label high value orders
df["payment_type"]=df["payment_method"].map({
    "Card":"Digital",
    "Wallet":"Digital",
    "COD":"Cash"
})
```

```python
#verify
df["payment_type"].unique() # check unique value
```

```python
df[["payment_method","payment_type"]].head() # compare before vs after
```

```python
df.groupby("payment_type").size() # business sanity check
```

```python
# apply() - create revenue category
def revenue_category(row):
    if(row)["revenue"]>500:
        return "High"
    elif row["revenue"]>=200:
        return "Medium"
    else:
        return "Low"
df["revenue_category"]= df.apply(revenue_category,axis=1)
```

```python
# verify
df["revenue_category"].value_counts() # check cetegory
```

```python
df[["revenue","revenue_category"]].sample(10) # validate logoc manually
```

```python
# pipe() - method chaining
def add_profit(data):
    data["profit"]=data["revenue"]*0.20
    return data
def add_net_revenue(data):
    data["net_revenue"]=data["revenue"]-data["profit"]
    return data
df=(
df
.pipe(add_profit)
.pipe(add_net_revenue)
)
```

```python
# verify
df[["revenue","profit","net_revenue"]].head()  # check new column
```
