```python
# class object , method, attribute 
class car:
    def __init__(self,color):
        self.color = color
    def start(self):
        print(f"{self.color} car started")
my_car = car("Black")
my_car.start()
```

```python
# defining class and creating object 
class car:
    def drive(self):
        print("Car is moving!")
myCar = car()
myCar.drive()
```

```python
class car:
    color = "Black"
    def drive(self):
        print(f"{self.color} car is moving")
myCar = car()
myCar.drive()
```

```python
# we can also edit object
class car:
    color = "Black"
    def drive(self):
        print(f"{self.color} car is moving")
    def setColor(self,newColor):
        self.color = newColor
myCar=car()
myCar.setColor("Red")
myCar.drive()
```

```python
# the __init__() contructor method we call it for setting default value.
class dog:
    def __init__(self):
        self.name = "Buddy"
    def bark(self):
        print(f"{self.name} says woof!")
myDog= dog()
myDog.bark()
```

```python
class car:
    def __init__(self,brand,color):
        self.brand=brand
        self.color=color
myCar=car("Frari","Black")
print(myCar.brand)
print(myCar.color)
myCar.color="Blue"
myCar2=car("Civic","White")
print(myCar2.brand)
print(myCar2.color)
print(myCar.color)
```

```python
# instant variable and instant method, instant variable declare within method and represent as self.
class student:
    def __init__(self,name,age,grade):
        self.name=name
        self.age=age
        self.grade=grade
    def display_info(self):
        print(f"Name : {self.name}")
        print(f"Age : {self.age}")
        print(f"Grade : {self.grade}")
    def is_Pass(self):
        if self.grade>50:
            print("Congradulation! You passed.")
        else:
            print("OOPS! You fail.")
myStudent=student("Abdullah",27,67)
myStudent.display_info()
myStudent.is_Pass()
```
