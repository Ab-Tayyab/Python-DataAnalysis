```python
# why seaborn?
# Cleaner visuals with minimal code
# built-in themes & color palettes
# high-level statistical plots
# ideal for expolaratory data analysis (EDA) in AI / ML pipelines
```

```python
# key feature of seaborn
# high-level plotting functions (sns.displot, scatter plots)
# automatic handling of pandas dataframs
# built-in themes: darkgrid, whitegrid, dark, white, ticks
# seamless integration with numpy, pandas, matplotlib
```

```python
# seaborn offers visual for:
# distribution plots, histogram, KDE
# categorical plots, bar, box, violin
# regression / relationship scatter plots with trend line
# matrix plots, heatmaps, pairgrid maps
# provide quick visual mapping for understanding
```

```python
# plots level in seaborn
# figure- level plots: relplot(),catplot(),displot() manage facets & multiple axes automatically
# axes-level plots: scatterplot(),boxplot(),histplot(), single subplot control , integrates with matplotlib axes
# themes: sns.set_style("darkgrid",sns.set_context("talk") controls background, grid, font, sizes for presenation
```

```python
# load the healthcare dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("./files/healthcare_patient_analytics_seaborn.csv")
```

```python
df = df.sample(500)
#axes-level example (scatterplot)
sns.scatterplot(
    data=df,
    x="treatment_cost",
    y="recovery_score",
    hue="visit_type"
)
plt.title("Treatment cost vs recovery score")
plt.show()
```

```python
# figure- level example (relplot)
sns.relplot(
    data=df,
    x="treatment_cost",
    y="recovery_score",
    hue="visit_type",
    col="gender"
)
```

```python
# apply seaborn theme
sns.set_theme(style="whitegrid") # always set the theme before plotting
```

```python
# recreating a matplotlib plot in seaborn
# matplotlib
plt.scatter(df["treatment_cost"],df["recovery_score"])
plt.xlabel("Treatment cost")
plt.ylabel("Recovery Score")
```

```python
# seaborn
sns.scatterplot(
    data=df,
    x="treatment_cost",
    y="recovery_score"
)
```
