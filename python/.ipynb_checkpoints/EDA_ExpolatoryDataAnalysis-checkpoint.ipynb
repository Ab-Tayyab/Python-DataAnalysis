```python
# Desctiptive stats & data distributions 
# key functions for descriptive statistics
# df.describe()  provide statistic data
# df['col'].value_counts()  provide number of value 
# df.quantile(q)  divide dataset into 4 equal quantile, every quantile have 25% data. 
# skewness() left or right skewed
# kurtosis analysis
```

```python
df = pd.read_csv("./files/ecommerce_sales_analytics_5000.csv",parse_dates=["order_date"])
```

```python
df.shape  # verify dataset before analysis
```

```python
# core decriptive statistics
df.describe()
```

```python
df[["unit_price","revenue"]].describe() # for revenue values
```

```python
# quantiles
df["revenue"].quantile([0.25,0.5,0.75])
```

```python
# identifying skewness
df["revenue"].mean(),df["revenue"].median()  # in this case mean is slightly right sid of the median
```

```python
df["revenue"].skew() # positive value right skew: if value +ve dataset skewed right side and if value -ve its mean dataset left side skewed
```

```python
# category -level descriptive statistics. 
df.groupby("product_category")["revenue"].describe() # do different product categories behave differently?
```

```python
df.groupby("product_category")["revenue"].mean() # find highest category
```

```python
 # using IQR to understand data spread
Q1=df["revenue"].quantile(0.25)
Q3=df["revenue"].quantile(0.75)
IQR = Q3-Q1
Q1,Q3,IQR
# helps understand normal range
# prepares us for outlier detection
# in this case IQR tells us how much data spread between Q1 and Q3 range
```

```python
(Q3>Q1) and (IQR>0) # check data is Q3 is greater then to Q1 and IQR is positive, this tells us data is valid
```

```python
# working with datetime & time series data 
# in real world data is provide by time series data 
df = pd.read_csv("./files/ecommerce_sales_analytics_5000.csv",parse_dates=["order_date"])
```

```python
df = df.set_index("order_date") #we treat data as time series data
```

```python
df.index
```

```python
# monthly revenue analysis
monthly_revenue = df["revenue"].resample("ME").sum()
monthly_revenue # end of month last date monthly revenue
```

```python
monthly_revenue.pct_change()  #we convert revenue into percentage
```

```python
monthly_revenue.index.is_monotonic_increasing # verify is index is monotonocaly increasing or is jump find like increase and sudenly decrease 
```

```python
print(monthly_revenue.sum(),df["revenue"].sum())
monthly_revenue.sum()<=df["revenue"].sum() # check end of month sum is greater then totao reveue
```
