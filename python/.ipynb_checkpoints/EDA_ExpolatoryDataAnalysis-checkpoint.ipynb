```python
# outlier detection using statistical rules
# outlier are data points that fall outside tha normal range, representing extreme values that can distort the mean & median of dataset
# IQR method
# z-scroe method
# domain driven thresholds
```

```python
df = pd.read_csv("./files/ecommerce_sales_analytics_5000.csv",parse_dates=["order_date"])
```

```python
# IQR method 
Q1 = df["revenue"].quantile(0.25)
Q3 = df["revenue"].quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
outliers_iqr = df[(df["revenue"]<lower_bound) | (df["revenue"]>upper_bound)]
# IQR (Interquatile range): Measure the middle 50% of data. Robust for showed distibutions (like revenue)
```

```python
outliers_iqr.shape # verification
```

```python
outliers_iqr[["revenue","product_category"]].head() #inspect
```

```python
df = df[(df["revenue"]>=lower_bound) & (df["revenue"]<=upper_bound)]
df.shape
```

```python
# z-score method
```

```python
from scipy.stats import zscore
df = pd.read_csv("./files/ecommerce_sales_analytics_5000.csv",parse_dates=["order_date"])
df["revenue_z"] = zscore(df["revenue"])
outliers_z = df[df["revenue_z"].abs()>3]
```

```python
# Z-score shows how many standard deviations a value is from the mean
# rule:[z]>3->outliers
```

```python
outliers_z.shape # verification
```

```python
df = df[df["revenue_z"].abs()<=3]
df.shape
```

```python
# time series data we use iqr method and for other data we use z-score
```

```python
# Correlation & Co-variance Analysis
# Measure teh relationship between numerrical variables
# if two variables increase together or decrease together is called positive correlation
# if one increase and other decrease is called negative correlation.  
```

```python
# covariance work with absolute value and it is positive or negative
df = pd.read_csv("./files/ecommerce_sales_analytics_5000.csv",parse_dates=["order_date"])
numeric_df = df[["quantity","unit_price","discount","delivery_days","revenue"]]
# covariance matrix
print("\nCovariance:\n",numeric_df.cov())
```

```python
# coorelation matrix
# linear correlation betwwen deffirent negative or positve value in percentage
print("\nCorrelation:\n",numeric_df.corr())
```

```python
numeric_df.corr()["revenue"].sort_values(ascending=False) # focus on revenue relations
```

```python
numeric_df.corr().values.min(),numeric_df.corr().values.max() # veification
# expected: minmam>= -1 maximum<= +1
```

```python
# correlation & covariance is not cosation, cosation mean which specific one variable change due to other
```
