```python
# instant variable and instant method, instant variable declare within method and represent as self.
class student:
    def __init__(self,name,age,grade):
        self.name=name
        self.age=age
        self.grade=grade
    def display_info(self):
        print(f"Name : {self.name}")
        print(f"Age : {self.age}")
        print(f"Grade : {self.grade}")
    def is_eligible(self):
        if self.age>=16:
            print("Congradulation! You are eligible for admission .")
        else:
            print("OOPS! You are not eligible for admission. ")
student1=student("Abdullah",16,"10th")
student2=student("Tayyab",14,"8th")
```

```python
# calling method 
student1.display_info()
student1.is_eligible()
```

```python
# Encapsulation public & private  in this method we hide attribute like private attribute within class, we cannot access directly outside the class, if we want to access attribute we get them by get method.
class student:
    def __init__(self,name,age):
        self.__age=age
        self.__name=name
    def get_data(self):
        return self.__age
        return self.__name
student = student("Abdullah",16)
print(student.get_data())
```

```python
class student:
    @property
    def grade(self):
        return self.__grade
    def __init__(self,name,grade):
        self.__grade=grade
    def get_grade(self):
        return self.__grade
    def set_grade(self,grade):
        if 0<=grade<=100:
            self.__grade=grade
        else:
            print("Invalid grade!")
student = student("Tayyab",90)
student.set_grade(95)
student.get_grade()
```

```python
# getter and setter 
class isPrime:
    def __init__(self,value):
        self.__value = value
    def get_value(self):
        return self.__value
    def set_value(self,value):
        count = 0
        hValue=value//2
        for i in range(2,hValue+1):
            if(value%i==0):
                count +=1
        if(value>1 and count==0):
            self.__value = "Prime Number"
        else:
            self.__value = "Not Prime Number"

prime = isPrime(20)
for i in range(1,21):
    prime.set_value(i)
    print(i," -> ",prime.get_value())
# prime.set_value(7)
# prime.get_value()
```

```python
# method overriding  use and modified parent class function into child class
class animal:
    def speak(self,a,b):
        print(a+b)
class dog(animal):
    def speak(self,a,b): #redefine function 
        print(a-b)
parentDog=animal()
parentDog.speak(10,20)
childDog = dog()
childDog.speak(10,20)
```

```python
# poly morphism - overriding % over loading. polymorphism mean multiple , we use diffrent class at a time and call a same function
class Cat:
    def speak(self):
        print("Meo Meo!")
class Dog:
    def speak(self):
        print("Whoo Whoo")
for pets in [Cat(),Dog()]:
    pets.speak()
```

```python
# overriding in polymorphism 
# poly morphism - overriding % over loading. polymorphism mean multiple , we use diffrent class at a time and call a same function
class vehicle:
    def move(self):
        print("Vehicle is moving")
class Car(vehicle):
    def move(self):
        print("Car us driving")
car1 = Car()
car1.move()
```

```python
# method overloading mean we create one function and apply diffrent method as required or diffrent number of arguments
class Calculator:
    def add(self,a,b=0,c=0):
        return a+b+c
cal = Calculator()
print(cal.add(10))
print(cal.add(10,20))
print(cal.add(10,20,30))
```

```python
# speciall method of class that python run by default, it is create by __str__ like constructor
class Book:
    def __init__(self,title,author):
        self.title = title
        self.author = author
    def __str__(self):
        return f"'{self.title}' by {self.author}"
b=Book("Quran","Allah")
print(b)
```

```python
class playList:
    def __init__(self,songs):
        self.songs=songs
    def __len__(self):
        return len(self.songs)
    def __add__(self,other):
        return playList(self.songs + other.songs)
p1 = playList(["Song 1","Song 2"])
p2 = playList(["Song 3"])
print(len(p1))
print(len(p2))
print (len(p1+p2))
```
