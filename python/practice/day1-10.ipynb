```python
# What is the sys.maxsize value in Python, and how does Python handle integers larger than it?
# sys.maxsize is the largest positive integer a Python index-sized integer can have on your system. 
# It usually reflects the maximum size of containers like lists or strings.
# On a 64-bit system, it is typically: 9223372036854775807  # (2^63 - 1)
# On a 32-bit system, it is:2147483647  # (2^31 - 1)

import sys
print(sys.maxsize)

# 🔹 Does it limit Python integers?
# No—this is where Python is different from many other languages.
# Python integers (int) are unbounded, meaning:
# They can grow as large as your system’s memory allows.
# They are not limited by sys.maxsize.

# 🔹 How Python handles very large integers
# When numbers exceed typical fixed-size limits:
# Python automatically switches to using more memory.
# Internally, it represents integers using multiple machine words (arbitrary-precision arithmetic).

x = 9223372036854775807  # sys.maxsize on 64-bit
y = x + 1

print(y)
# No overflow occurs—Python just keeps going.
```

```python
# Write code demonstrating the difference between == (equality) and is (identity) for small integers vs large integers

# In Python, == checks value equality, while is checks object identity (whether both variables point to the same object in memory).
# Small integers are often cached (interned), while large integers usually are not—this creates a visible difference.

# Small integers
a = 10
b = 10

print("Small integers:")
print("a == b:", a == b)  # True (same value)
print("a is b:", a is b)  # True (same object due to interning)

# Large integers
x = 1000
y = 1000

print("\nLarge integers:")
print("x == y:", x == y)  # True (same value)
print("x is y:", x is y)  # Usually False (different objects)

# Forcing creation of separate objects explicitly
m = int("1000")
n = int("1000")

print("\nLarge integers (explicit objects):")
print("m == n:", m == n)  # True
print("m is n:", m is n)  # False
```

```python
# What is integer interning in Python? Write code to demonstrate when two variables share the same object in memory.

# Integer interning is an optimization where Python reuses existing integer objects instead of creating new ones.
# Python automatically interns (caches) small integers, typically in the range -5 to 256.
# When you assign variables within this range, they often point to the same memory object.
# For integers outside this range, Python usually creates separate objects.

# Small integers (interned)
a = 100
b = 100

print("Small integers:")
print("a == b:", a == b)  # True (same value)
print("a is b:", a is b)  # True (same object)

# Large integers (not always interned)
x = 1000
y = 1000

print("\nLarge integers:")
print("x == y:", x == y)  # True (same value)
print("x is y:", x is y)  # Usually False (different objects)

# Force separate objects
m = int("1000")
n = int("1000")

print("\nExplicitly created integers:")
print("m == n:", m == n)  # True
print("m is n:", m is n)  # False

# Checking memory addresses
print("\nMemory addresses:")
print("id(a):", id(a))
print("id(b):", id(b))
print("id(x):", id(x))
print("id(y):", id(y))
print("id(m):", id(m))
print("id(n):", id(n))
```

```python
# Write code that uses the bytes and bytearray types to store the string 'Python' in binary form.

# Original string
text = "Python"

# Convert to bytes (immutable binary data)
byte_data = bytes(text, "utf-8")
print("Bytes:", byte_data)

# Optional: demonstrate that bytes is immutable
byte_data[0] = ord('p')  # change 'P' to 'p'
print("Modified bytearray:", byte_data)

# Convert to bytearray (mutable binary data)
byte_array = bytearray(text, "utf-8")
print("Bytearray:", byte_array)

# Optional: demonstrate that bytearray is mutable
byte_array[0] = ord('p')  # change 'P' to 'p'
print("Modified bytearray:", byte_array)


# Explanation:
# bytes() creates an immutable sequence of bytes.
# bytearray() creates a mutable sequence of bytes.
# "utf-8" encoding converts the string into its binary representation.
```
