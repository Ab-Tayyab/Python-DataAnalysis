```python
# list are ordered and mutable (change able), looks like arrays and use square brackets and have 
# all kind of data.
data=["Abdullah",28,True,3.14]
print("Get element by index: ",[2])
data[2]=False
print("After update specific element: ",data) 
removed= data.pop()
print("After remove last element: ",data)
print("Removed element",removed)
data.insert(1,"Tayyab")
print("Insert element at specific position: ",data)
data.remove("Tayyab")
print("Remove any element: ",data)
data.append("Tayyab")
print("Add element at last of the list: ",data)
data.reverse()
print("After reverse list: ",data)
print("Access last element: ",data[-1])
# loop iteration 
for item in data:
    print(item)
```

```python
# tuples are ordered and immutable (not change), use curly brackets and hav all type of data.
data=("Abdullah",25,3.14,True,25)
print(data)
print("Access first alement: ",data[0]," and access last element of tuple: ",data[-1])
print("Get specific element index number: ",data.index(3.14))
print("Get total number of same element: ",data.count(25))
# loop over tuple 
for item in data:
    print(item)


```

```python
# sets are unique that have not duplicate value thats why sets work fast and use { } brackets
data = {1,2,3,4,4,5}
print("Duplicate value auto remove: ",data)
data.add(7)
print("Add one element at last index: ",data)
data.update("Abdullah")
print("Add multiple element at last but it works only for string not integer or other",data)
data.remove(5)
print("Remove element at specific index: ",data)
removed = data.pop()
print("Remove first element: ",data, "Removed value is: ",removed)
a={1,2,3}
b={3,4,5}
print("A union B: ",a.union(b))
print("A intersection B: ",a.intersection(b))
print("A difference B: ",a.difference(b))
print("A symmetric_difference B: ",a.symmetric_difference(b))
# loops in sets 
for item in data:
    print(item)
```

```python
# Dictionary are like objects uses key value pair and use { } 
data={
    "name":"abdullah",
    "age":28,
    "course":"Ai automation",
    "isComplete":False
}
print(data)
print("Get specific element by key: ",data["age"])
print("Get specific element by key: ",data.get("isComplete"))
data["City"]="Lahore"
print("Add value at last: ", data)
data["age"]=27
print("Uppdate value by key: ",data)
removed = data.pop("City")
print("Element remove using key: ",data," an removed value is: ",removed)
# loop on disctionary 
for key,value in data.items():
    print(key," : ",value)

```
