```python
# load the healthcare dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("./files/healthcare_patient_analytics_seaborn.csv")
```

```python
df = df.sample(500)
#axes-level example (scatterplot)
sns.scatterplot(
    data=df,
    x="treatment_cost",
    y="recovery_score",
    hue="visit_type"
)
plt.title("Treatment cost vs recovery score")
plt.show()
```

```python
# figure- level example (relplot)
sns.relplot(
    data=df,
    x="treatment_cost",
    y="recovery_score",
    hue="visit_type",
    col="gender"
)
```

```python
# apply seaborn theme
sns.set_theme(style="whitegrid") # always set the theme before plotting
```

```python
# recreating a matplotlib plot in seaborn
# matplotlib
plt.scatter(df["treatment_cost"],df["recovery_score"])
plt.xlabel("Treatment cost")
plt.ylabel("Recovery Score")
```

```python
# seaborn
sns.scatterplot(
    data=df,
    x="treatment_cost",
    y="recovery_score"
)
```

```python
# distribution with seaborn
# sns.hisplot()-> histograms with multiple, stacked or layered groups
# sns.kdeplot()-> kernal density estimate
# sns.ecdfplot() -> empirical cumulative distribution function
# common argument
# multiple-> 'layer', 'stack','dodge'
# stat -> 'count','density','probability'ideal for understanding data shape, spread, & skewness
```

```python
# histoplot
# senario - treatment cost distribution
sns.histplot(
    data = df,
    x = "treatment_cost",
    bins=30
)
plt.title("Distribution of treament cost")
plt.show()
```

```python
# comparing distribution across categories
# senario - routine vs emergency visits
sns.histplot(
    data = df,
    x = "readmission_risk",
    hue="visit_type",
    multiple="stack",
    bins=30
)
plt.title("Re-admission by visit type")
plt.show()
```

```python
# kdeplot - provide a smooth estimate of the distribution
sns.kdeplot(
    data=df,
    x="recovery_score",
    hue = "visit_type",
    fill=True
)
plt.title("KDE of recovery score by visit type")
plt.show()
```

```python
# ECDF - shows cumulative proportion
sns.ecdfplot(
    data=df,
    x="length_of_stay_days",
    hue="visit_type"
)
plt.title("ECDF of length of stay Days")
plt.show()
```

```python
# Categorical data visualization with seaborn
# families of categorical plots
# aggregate (central tendency graphs) - bar plot count plot
# distribution plot - box plot, violin plot, swarm plot, strip plot
```

```python
df = df.sample(500)
# barplot vs countplot
# - countplot:
sns.countplot(
    data=df,
    x="treatment_type"
)
plt.title("Number of visits by treatment type")
plt.show()
```

```python
# barplot vs countplot
# bar plot
sns.barplot(
    data=df,
    x="department",
    y="treatment_cost"
)
plt.xticks(rotation=30)
plt.title("Average treatment cost by department")
plt.show()
```

```python
# distribution based categorical plots
# boxplot - shows median, quartiles, and outliers
sns.boxplot(
    data=df,
    x="department",
    y="recovery_score"
)
plt.xticks(rotation=30)
plt.show()
```

```python
# distribution based categorical plots
# swarm / strip plot - shows individual data points
sns.stripplot(
    data=df,
    x="department",
    y="recovery_score",
    jitter=True
)
plt.xticks(rotation=30)
plt.show()
```

```python
# distribution based categorical plots
# violinplot - shows full distribution shape
sns.violinplot(
    data=df,
    x="department",
    y="recovery_score"
)
plt.xticks(rotation=30)
plt.show()
```

```python
# side by side comaprison (box + strip)
sns.boxplot(
    data=df,
    x="department",
    y="recovery_score"
)
sns.stripplot(
    data=df,
    x="department",
    y="recovery_score",
    color="black",
    alpha=0.4
)
plt.xticks(rotation=30)
plt.show()
    
```

```python
# heatmaps & correlation visualization 
numeric_df=df[
    ["length_of_stay_days","treatment_cost","recovery_score","readmission_risk"]
    ]
```

```python
# correlation matrix
corr_matrix = numeric_df.corr()
corr_matrix
```

```python
# heatmaps basic visualization
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm"
)
plt.title("correlation heatmap (healthcare data")
plt.show()
```

```python
# masking upper triangle
mask = np.triu(np.ones_like(corr_matrix,dtype=bool))
sns.heatmap(
    corr_matrix,
    mask = mask,
        annot=True,
    cmap="coolwarm"
)
plt.title("correlation heatmap (Masked)")
plt.show()
```
