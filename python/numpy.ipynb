```python
or_arr = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])
```

```python
print(or_arr[0,1]) # row 0, col 1
print(or_arr[2,2]) # row 2 col 2
print(or_arr[-3,-3]) # row 1 col 1
```

```python
# general form : or_arr[start:stop:step]
print(or_arr[0:2,1:3]) # rows 0-1, cols 1-2
```

```python
print(or_arr[::2,::2]) # every 2nd row and 2nd column
```

```python
slice_arr=or_arr[0,:2] # view not a copy
slice_arr[0]=999
```

```python
print(slice_arr[0]) # original changes
```

```python
copy_arr= or_arr[1].copy() # independent copy
copy_arr[0]=555
```

```python
print(or_arr[1]) # original unchanged
print(copy_arr[1])
```

```python
# simulate 100k coin flips
rng = np.random.default_rng(seed=43) # seed use for same random number sequence generator witout seed some times head increase and some time tial increase
flips = rng.integers(0,2,size=100000) # 0-tail, 1=heads
prob_heads = np.mean(flips)
print("Estimated probability of heads: ", prob_heads)
```
