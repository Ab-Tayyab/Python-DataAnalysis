```python
import numpy as np
import time
```

```python
arr = np.array([[1,2,3],[44,5,6]])
print(arr.ndim) # for check dimensions
print(arr.dtype) # for check array type like int, float etc
print(arr.shape) # for check rows and column
```

```python
# check execution time of python list 
py_list = list(range(1000000))
arr=[]
start_time = time.time()
for i in py_list:
    arr.append(i+5)
end_time = time.time()
print(f"Python list execute time is: {end_time - start_time}")
```

```python
# check execution time of numpy array
np_array = np.array(1000000)
start_time = time.time()
np_result = np_array+5
end_time = time.time()
print(f"Numpy array execute time is: {end_time - start_time}")
```
