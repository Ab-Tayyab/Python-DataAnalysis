```python
# we use context manager for text handling, file open specfic time after function excecute and close autometically 
with open("./files/myFile.txt","r") as file:
    data=file.read()
    print(data)
```

```python
# handling exceptions during file operation 
with open("./files/missingFile.txt") as file:
    data=file.read()
    print(data)
```

```python
try:
    with open("./files/missingFile.txt") as file:
        data=file.read()
        print(data)
except:
    print("File not found")
```

```python
try:
    with open("./files/missingFile.txt") as file:
        print(file.read())
except FileNotFoundError:
    with open("./files/missingFile.txt","w") as file:
        file.write("Exception hanling")
    with open("./files/missingFile.txt") as file:
        print(file.read())
```

```python
# try except else finally 
try:
    num = int(input("Enter number: "))
    print(10/num)
except ZeroDivisionError:
    print("You can not divide a number by zero")
except ValueError:
    print("You can not divide a number by character")
except Exception as e:
    print("Unexpected error accur: ",e)
```

```python
try:
    print("Code is running...")
except:
    print("Error found")
else:
    print("No error found")
finally:
    print("Done!")
```

```python
# raise error, wo also create self generate error using raise keyword
age = -2
if age<0:
    raise ValueError("Age cannot be negative: ")
```

```python
# creating custom exception 
class invalidAgeError(Exception):
    pass
```

```python
age = int(input("Enter age: "))
if age<0:
    raise invalidAgeError(f"Age cannot be nagative, {age}")
```

```python
try:
    age = int(input("Enter age: "))
    if age<0:
        raise invalidAgeError(f"Age cannot be nagative, {age}")
except invalidAgeError as e:
    print(e)
```
